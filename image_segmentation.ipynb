```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
x=np.array(Image.open('cow.jpg'))
plt.imshow(x,plt.cm.gray)
plt.show()
```

```python
[p,q,t]=x.shape
x=x/255.0

```

```python
choose=np.zeros([p,q])
choose[125:130,100:200]=1.0
plt.imshow(choose,plt.cm.gray)
```

```python
alpha=1
beta=0.5
y=choose
myvalue=np.zeros([p,q])
for l in range(1,p-1):
    for k in range(1,q-1):
        Energy1=abs(x[l,k]-x[125:130,100:200])+(abs(y[l,k]-y[l+1,k])+abs(y[l,k]-y[l-1,k])+abs(y[l,k]-y[l,k-1])+abs(y[l,k]-y[l,k+1]))

        if np.sum(Energy1)<150:
            myvalue[l,k]=1
        else:
            myvalue[l,k]=0
            
```

```python
np.sum(Energy1),np.sum(np.sum(np.sum(Energy1,axis=2),axis=1))
```

```python
plt.imshow(myvalue,plt.cm.gray)
plt.show()
```
