```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
xx=np.array(Image.open('im.jpg').convert("L"))
plt.imshow(xx,plt.cm.gray)
plt.show()
```

```python
[m,n]=xx.shape
Final_image=np.zeros([m,n])
y=xx
x=y
for it in range(50):
    for i in range(1,m-1):
        for j in range(1,n-1):
            new_value=np.array([25,50,100,200]) #all possible state we may think to be present
            Energy=abs(new_value-x[i,j])+((abs(x[i,j+1]-new_value)+abs(x[i,j-1]-new_value)+abs(x[i-1,j]-new_value)+abs(x[i+1,j]-new_value)+abs(x[i-1,j-1]-new_value)+abs(x[i-1,j+1]-new_value)+abs(x[i+1,j-1]-new_value)+abs(x[i+1,j+1]-new_value)))
            select=np.argmin(Energy)
            Final_image[i,j]=new_value[select]
    x=Final_image
```

```python
plt.imshow(x,plt.cm.gray)
```
